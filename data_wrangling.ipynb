### Reshaping long to wide

```python
air_quality = pd.read_csv("air_quality_long.csv", index_col="date.utc", parse_dates=True)
air_quality
```

Use 'pivot' to move values from a column into columns of a new frame:

```python
air_quality[air_quality['parameter'] == 'no2'].pivot(columns="location", values="value")
```

Note that a single value for each index/column combination is required (hence why we're subsetting to no2 data)

Suppose we want the mean concentrations for no2 and pm2.5 and in each of the stations in table form.

```python
air_quality.pivot_table(values="value", index="location", columns="parameter", aggfunc="mean")
```

In the case of pivot(), the data is only rearranged. When multiple values need to be aggregated (in this specific case, the values on different time steps), pivot_table() can be used, providing an aggregation function (e.g. mean) on how to combine these values.

### Reshaping wide to long

```python
no2_pivoted = air_quality[air_quality['parameter'] == 'no2'].pivot(columns="location", values="value").reset_index()
no2_pivoted
```

Collect air quality no2 measurements in a single column (long format):

```python
no2_pivoted.melt(id_vars="date.utc")
```

The pandas.melt() method on a DataFrame converts the data table from wide format to long format. The column headers become the variable names in a newly created column.

Additional parameters have the following effects:

    value_vars defines which columns to melt together

    value_name provides a custom column name for the values column instead of the default column name value

    var_name provides a custom column name for the column collecting the column header names. Otherwise it takes the index name or a default variable

Hence, the arguments value_name and var_name are just user-defined names for the two generated columns. The columns to melt are defined by id_vars and value_vars.

```python
no2_long = no2_pivoted.melt(id_vars="date.utc", value_vars=["BETR801", "FR04014", "London Westminster"],
                            value_name="NO_2", var_name="id_location",)
no2_long
```

```python
sns.lineplot(no2_long,x='date.utc',y='NO_2',hue='id_location')
```

# Case study in data wrangling

## Yeast Genomics
[Smith and Kruglyak (2008)](https://journals.plos.org/plosbiology/article?id=10.1371/journal.pbio.0060083) is a study that measured 2820 genotypes in 109 yeast F1 segregants from a cross between parental lines BY and RM.
They also measured gene expression on 4482 genes in each of these segregants when growing in two different Carbon sources, glucose and ethanol.

```python
raw_data = pd.read_csv('Dataset_S1.csv')
raw_data
```

```python
raw_data = pd.read_csv('Dataset_S1.csv',header=None,index_col=0).T
raw_data
```

We're getting there, but the gene expression values (everything after the 'Dye' column) are not formatted as numbers. To see this:

```python
raw_data.iloc[:,5:].values #':' select all value ; '5:'select all columns start from index 5(columns 6)
```

Note the values all appear as strings. Let's convert these to numbers

```python
raw_data.iloc[:,5:] = raw_data.iloc[:,5:].apply(pd.to_numeric)
raw_data.iloc[:,5:].values
```

Now let's take the median value across replicates:

```python
median_expression = raw_data.drop(columns=['Dye','Replicate']).groupby(['Name','Strain','Cond']).median().apply(pd.to_numeric)
median_expression
```

```python
summary = (2**median_expression).describe()
summary
```

Let's just make sure every gene has the same number of observations:

```python
summary.loc['count'].describe()
```

We can have a look at the coefficient of variation

```python
summary.loc['CoV'] = summary.loc['std']/summary.loc['mean']
sns.scatterplot(x=summary.loc['mean'], y=summary.loc['CoV'])
```

```python
def our_pca(X,center=True):
    if center:
        X = X.apply(lambda x: x-x.mean())
    U,sigma,Vt = np.linalg.svd(X.T,full_matrices=False)
    PCs = np.diag(sigma).dot(Vt)
    PVE = sigma**2/(sigma**2).sum()
    return PCs, U, PVE
```

```python
PCs,U,PVE = our_pca(median_expression)
df_PCs = pd.DataFrame(PCs.T)
df_PCs.columns = ['PC%d'%(i+1) for i in range(PCs.shape[0])]
```

```python
sns.scatterplot(PVE) # 24 dimension but mostly 3 
```

```python
sns.scatterplot(df_PCs,x='PC1',y='PC2') # dot=samples, have 24 samples,
```

```python
df_PCs['Strain'] = median_expression.index.get_level_values(1)
df_PCs['Cond'] = median_expression.index.get_level_values(2)

sns.scatterplot(df_PCs,x='PC1',y='PC2', hue='Strain', style='Cond')
```
